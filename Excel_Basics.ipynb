Sol1.Difference between Relative, Absolute, and Mixed Cell Referencing
    •	Relative Reference (e.g., A1)
      Changes automatically when you copy the formula to another cell.
      Example: If =A1+B1 is copied down, it becomes =A2+B2.
    •	Absolute Reference (e.g., $A$1)
      Stays fixed no matter where you copy the formula.
      Example: =$A$1+B1 = A1 will never change.
    •	Mixed Reference (e.g., $A1 or A$1)
      One part is fixed, the other changes:
      $A1 = Column fixed, row changes
      A$1 = Row fixed, column changes

Sol2.Formula to calculate total sales of Car and Bicycle only
    Assuming:
      Car price is in B2
      Bicycle price is in B3
    Formula:
      =SUM(B2:B3)

Sol3.Formula to calculate average sales where price >100 and <300
    Using the table:
    Price in column B
    Sales in column C
    Formula:
    =AVERAGEIFS(C2:C4, B2:B4, ">100", B2:B4, "<300")

Sol4. Count how many customer names are recorded
    Use COUNTA (counts non-empty cells):
    =COUNTA(A2:A100)

Sol5. Calculate the Total Sales for each row
    Use multiplication (e.g., Quantity * Price):=C2*D2

Sol6. Total sales of Notebooks in the North region only
    Use SUMIFS:
    =SUMIFS(E2:E100, B2:B100, "Notebook", C2:C100, "North")
    	E = Total Sales column
    	B = Product column
    	C = Region column

Sol7. Create a column chart showing total sales by product
    Steps:
    a)Select Product and Total Sales columns.
    b)Go to Insert → Chart.
    c)Choose Column Chart.
    d)Add title: "Total Sales by Product".

Sol8. Insert a line chart showing daily sales trend
Steps:
    a)Select Date and Total Sales columns.
    b)Go to Insert → Chart.
    c)Choose Line Chart.
    d)Add title: "Daily Sales Trend".
